Importing the Labraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading the CSV File

```python
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

Dropping the Columns that are not Needed

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.info()
```

Dropping Duplicates

```python
df.drop_duplicates(inplace = True)
df.shape
```

Cleaning the Rate Columns

```python
df['rate'].unique()
```

Removing NEW , - and /5 from Rate Column

```python
def ratehandle(value):
    if(value == 'NEW' or value =='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
df['rate'] = df['rate'].apply(ratehandle)
df['rate'].head()
```

cheacking the null value in RATE column

```python
df.rate.isnull().sum()
```

checking the remaing null values in other columns


```python
df.isnull().sum()
```

Filling Null Values in Rate column with mean

```python
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

Dropping Null values

```python
df.dropna(inplace=True)
df.head()
```

Renaming Columns approx_cost(for two people) and listed_in(type)

```python
df.rename(columns={'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'}, inplace = True)
df.head()
```

```python
df.location.unique()
```

```python
df['listed_in(city)'].unique()
```

Listed in(city) and Location, both are there, So keep one

```python
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df.head()
```

Removing from cost2plates columns

```python
df['Cost2plates'].unique()
```

```python
def commahandle(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
df['Cost2plates']=df['Cost2plates'].apply(commahandle)
df['Cost2plates'].unique()
```

```python
df.head()
```

Cleaning rest type column

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

Making Rest Types less than 1000 in frequency as others

```python
def rest_type_handle(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type'] = df['rest_type'].apply(rest_type_handle)
df['rest_type'].value_counts()
```

```python
df.head()
```

Cleaning Location Columns

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending=False)
location_lessthan300 = location[location<300]
```

Making location less than 300 in frequency as others

```python
def location_hanlde(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
df['location'] = df['location'].apply(location_hanlde)
df['location'].value_counts()
```

```python
df.head()
```

Cleaning Cuisines Columns

```python
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending=False)
cuisines_lessthan100 = cuisines[cuisines<100]
```

Making cuisines less than 100 in frequency as others

```python
def cuisines_handle(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
df['cuisines']=df['cuisines'].apply(cuisines_handle)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df['Type'].value_counts()
```
